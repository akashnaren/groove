```python
#import packages
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import DBSCAN
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('CreditCard_dataset.csv')

#drop customer ID column
data.drop(columns=['CUST_ID'], inplace=True)

#replace null values with zeros
data['MINIMUM_PAYMENTS'].fillna(0, inplace=True)

scaler = StandardScaler()
scale_data = scaler.fit_transform(data)
scale_data = pd.DataFrame(scale_data, columns=data.columns)
```

```python
import numpy as np
from scipy.spatial.distance import pdist, squareform
import matplotlib.pyplot as plt

# Calculate the pairwise Euclidean distances between all points
euclidean_distances = pdist(scale_data.values, metric='euclidean')

# Plot a histogram of the Euclidean distances
plt.figure(figsize=(10, 6))
plt.hist(euclidean_distances, bins=100, alpha=0.75)
plt.title('Histogram of Pairwise Euclidean Distances')
plt.xlabel('Euclidean Distance')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

```python
euclidean_distances = pdist(scale_data.values, metric='euclidean')

# Sort the distances in ascending order
sorted_distances = np.sort(euclidean_distances)

plt.figure(figsize=(10, 6))
plt.plot(sorted_distances)
plt.title('Sorted Pairwise Euclidean Distances Between All Points')
plt.xlabel('Pair Index')
plt.ylabel('Euclidean Distance')
plt.grid(True)
plt.show()
```
